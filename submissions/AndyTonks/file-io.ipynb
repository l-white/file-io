```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('spotify-2023.csv',encoding='latin1')
df.head()
```

## Challenges
# Write a script to identify the number of songs in the file.


```python
songcount = df['track_name'].count()
print(f'Number of songs in file:= {songcount}')
```

# Write a script that identify the number of songs in the key of E.

```python
keyE = df.loc[df['key'] == 'E']
cnt = keyE['track_name'].count()
print(f'Number of songs in the key of E = {cnt}')
print('Listed below song in Key of E')
keyE
```

# Count the occurrences of values in a specified column (e.g., artist names) and determine the most common value.

```python
x = df['artist(s)_name'].value_counts()
sns.set_theme('paper')
plt.figure(figsize=(10,6))
x = x[:20]
sns.barplot(x,color='g')
plt.title('Top 20 Artists')
plt.xticks(rotation =90)

x
```
