# Challenge 3
# Count the occurrences of values in a specified column (e.g., artist names) and determine the most common value.

```python
# Calculate the value counts for each artist
artist_counts = df['artist(s)_name'].value_counts()

# Find the maximum count
max_count = artist_counts.max()

# Filter the artists with the maximum count
max_artists = artist_counts[artist_counts == max_count]

# Print the artist name(s) with the maximum count
print(max_artists.index.tolist())
```
