```python
keys  = file['key']
count = 0 
for key in keys:
    if key == "E":
        count +=1

print(f"Number of Songs in the key of E is {count}")
```

```python
def analyze_column(col, data):
    print(f"Analysis for column name {col}")
    if data is not None and col in data.columns:
        value_counts = data[col].value_counts()
        print("Value Counts: ")
        print(value_counts.head())

        most_common_value = value_counts.idxmax()
        most_common_value_count = value_counts.max()
        print(f"\n The most common value in {col} is '{most_common_value_count}' ")
    else:
        print("Data could not be loaded or column name is incorrect")

columns_to_analyze = ['artist(s)_name', 'released_year']

for column in columns_to_analyze:
    analyze_column(column, file)

```
